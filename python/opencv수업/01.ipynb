```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arrange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color="#e35f62", marker='*', linewidth=2)
plt.plot(x, x**3, color="forestgreen", marker='^', markersize=9)
plt.xticks([0,1,2])
plt.yticks(np.arange(1,6))

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('image/01_lena.png', cv2.IMREAD_COLOR)

# opencv는 색채널이 bgr로 되어있어서 rgb로 변경해주는 코드
# image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# plt.imshow(image)

# 직접 red채널과 blue채널을 바꾸기
b, g, r = cv2.split(image)
new_img = cv2.merge([r,g,b])
plt.imshow(new_img)
```

```python
# imread를 활용해서 이미지 gray로 만들기

image = cv2.imread('image/01_lena.png', cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap='gray')
```

```python
# 이미지 이진화
image = cv2.imread('image/01_lena.png', cv2.IMREAD_COLOR)

_, thresholded_image = cv2.threshold(image, 100, 225, cv2.THRESH_BINARY_INV)

plt.imshow(thresholded_image)
```

```python
# yt-dlp로 유튜브 영상 다운로드하기
import youtube_dl
import os

output_dir = os.path.join('videos/','%(title)s.%(ext)s')

download_list = ["https://youtu.be/JleoAppaxi0?si=_5zhsW4QZ0rZ0kzm"]

ydl_opt = {
    'outtmlpl': output_dir,
    'format': 'bestvideo/best',
}

with youtube_dl.YoutubeDL(ydl_opt) as ydl:
    ydl.download(download_list)
    
print('다운로드 완료')
```

```python
# 실습2 - 픽셀값 변경

import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('image/01_lena.png')
image[200, :, :] = [0]
b, g, r = cv2.split(image)
new_img = cv2.merge([r, g, b])

plt.imshow(new_img)
plt.show()
```

```python
# 실습 - 선 그리기

import cv2
img = cv2.imread('image/01_lena.png')
cv2.rectangle(img, (180, 250), (350, 430), (0, 0, 255), 3)


b, g, r = cv2.split(img)
new_img = cv2.merge([r, g, b])

plt.imshow(new_img)
plt.show()
```

```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
import cv2

img = cv2.imread('image/01_lenna.png')
img = cv2.rectangle(img, (180, 35), (20, 200), (0, 0, 255), 3)

font = ImageFont.truetype("fonts/gulim.ttc", 30)
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)

draw.text((30, 50), '빙봉', font = font, fill = (255,0,0,0))
img_pil.show()
```
