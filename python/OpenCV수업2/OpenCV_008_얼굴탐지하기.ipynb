```python

import cv2
video_path = 'videos/01.MOV'
cap = cv2.VideoCapture(video_path)

while True:
    ch, frame = cap.read()
    if not ch:
        print('동영상 재생 끝 or 읽기 실패')
        break

    cv2.imshow('video', frame)

    key = cv2.waitKey(33)
    if key == ord('q'):
        print('사용자에 의해서 영상이 종료')
        break
    
cap.release()

cv2.destroyAllWindows()
```
